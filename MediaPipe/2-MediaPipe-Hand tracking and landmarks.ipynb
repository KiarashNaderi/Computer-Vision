# hand tracking

```python
import cv2
import mediapipe as mp

mphand=mp.solutions.hands
hand=mphand.Hands()

mpdraw=mp.solutions.drawing_utils
cap=cv2.VideoCapture(0) # address camera ra midahad, sefr yani webcam
while(True):
    ret,frame=cap.read()
    frame=cv2.flip(frame,1) # drost shodane jahate camera
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # tabdile bgr be rgb
    res=hand.process(imgrgb)
    if res.multi_hand_landmarks:
        for counter in res.multi_hand_landmarks:
            mpdraw.draw_landmarks(frame,counter,mphand.HAND_CONNECTIONS)
    
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# tracking one point(drawing circle)

```python
import cv2
import mediapipe as mp

mphands=mp.solutions.hands
hands=mphands.Hands()

mpdraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(0)
while(True):
    ret,frame=cap.read()
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    res=hands.process(imgrgb)
    
    if res.multi_hand_landmarks:
        for handlandmarks in res.multi_hand_landmarks:
            for count,loc in enumerate(handlandmarks.landmark):
                h,w,c=frame.shape
                x,y=int(loc.x*w),int(loc.y*h)
                #print(count,x,y)
                
                if count==4:
                    cv2.circle(frame,(x,y),17,(150,230,90),-1)
                    #frame,mokhatast,shoa,color,-1
                mpdraw.draw_landmarks(frame,handlandmarks,mphands.HAND_CONNECTIONS)
    cv2.imshow("video",frame)
    if cv2.waitKey(1) &0XFF==ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# counter of fingers

```python
import cv2
import mediapipe as mp

mphands=mp.solutions.hands
hands=mphands.Hands(max_num_hands=1)

mpdraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(0)

while (True):
    ret,frame=cap.read()
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    res=hands.process(imgrgb)
    
    if res.multi_hand_landmarks:
        for hand_landmarks in res.multi_hand_landmarks:
            mpdraw.draw_landmarks(frame,hand_landmarks,mphands.HAND_CONNECTIONS)
            
            finger_count=0
            if hand_landmarks.landmark[4].x>hand_landmarks.landmark[2].x:
                finger_count+=1    #finger_count=finger_count+1 
                
            if hand_landmarks.landmark[8].y<hand_landmarks.landmark[6].y:
                finger_count+=1    #finger_count=finger_count+1 
                
            if hand_landmarks.landmark[12].y<hand_landmarks.landmark[10].y:
                finger_count+=1    #finger_count=finger_count+1 
                
            if hand_landmarks.landmark[16].y<hand_landmarks.landmark[14].y:
                finger_count+=1    #finger_count=finger_count+1 
                
            if hand_landmarks.landmark[20].y<hand_landmarks.landmark[18].y:
                finger_count+=1    #finger_count=finger_count+1 
            
            cv2.putText(frame,f"Fingers:  {finger_count}",(15,30),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    
    
    
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) &0XFF==ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()    
```
