# rasme dayere(circle) be tedad angoshtane dast

```python
import cv2
import mediapipe as mp

mphands=mp.solutions.hands
hands=mphands.Hands(max_num_hands=1)

mpdraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(0) # address camera ra midahad, sefr yani webcam
while(True):
    ret,frame=cap.read()
    frame=cv2.flip(frame,1) # drost shodane jahate camera
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # tabdile bgr be rgb
    res=hands.process(imgrgb)
    
    if res.multi_hand_landmarks:
        for hand_landmarks in res.multi_hand_landmarks:
            mpdraw.draw_landmarks(frame,hand_landmarks,mphands.HAND_CONNECTIONS)
            
        finger_count=0
        
        if hand_landmarks.landmark[4].x<hand_landmarks.landmark[2].x:
            finger_count+=1
            
        if hand_landmarks.landmark[8].y<hand_landmarks.landmark[6].y:
            finger_count+=1
            
        if hand_landmarks.landmark[12].y<hand_landmarks.landmark[10].y:
            finger_count+=1
            
        if hand_landmarks.landmark[16].y<hand_landmarks.landmark[14].y:
            finger_count+=1
        
        if hand_landmarks.landmark[20].y<hand_landmarks.landmark[18].y:
            finger_count+=1
            
        cv2.putText(frame,f" Fingers: {finger_count}",(15,30),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
        
        for i in range(finger_count):
            cv2.circle(frame,(50,50+50*i),10,(0,0,255),cv2.FILLED)
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# pose estimation

```python
import cv2
import mediapipe as mp

pose=mp.solutions.pose
mpbody=pose.Pose()


mpdraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(r"C:\Users\ASUS\Desktop\exercise01.mp4") # address camera ra midahad, sefr yani webcam
while(True):
    ret,frame=cap.read()
    frame=cv2.resize(frame,(600,400))
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # tabdile bgr be rgb
    res=mpbody.process(imgrgb)
    
    if res.pose_landmarks:
        mpdraw.draw_landmarks(frame,res.pose_landmarks,pose.POSE_CONNECTIONS)
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# Holistic detection==> jame hameye method ha ba ham

```python
import cv2
import mediapipe as mp

holis=mp.solutions.holistic
mpholi=holis.Holistic()


mpdraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(0) # address camera ra midahad, sefr yani webcam
while(True):
    ret,frame=cap.read()
    frame=cv2.flip(frame,1) # drost shodane jahate camera
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # tabdile bgr be rgb
    
    res=mpholi.process(imgrgb)
    
    mpdraw.draw_landmarks(frame,res.face_landmarks,holis.FACEMESH_CONTOURS)
    mpdraw.draw_landmarks(frame,res.left_hand_landmarks,holis.HAND_CONNECTIONS)
    mpdraw.draw_landmarks(frame,res.right_hand_landmarks,holis.HAND_CONNECTIONS)
    mpdraw.draw_landmarks(frame,res.pose_landmarks,holis.POSE_CONNECTIONS)
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
           
cap.release()
cv2.destroyAllWindows()
```

# dast nazan be mohat

```python
import cv2
import mediapipe as mp

mp_hands=mp.solutions.hands
hands=mp_hands.Hands(max_num_hands=1,min_detection_confidence=0.5,min_tracking_confidence=0.5)


cap=cv2.VideoCapture(0) # address camera ra midahad, sefr yani webcam


while(True):
    ret,frame=cap.read()
    frame=cv2.flip(frame,1) # drost shodane jahate camera
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # tabdile bgr be rgb
    res=hands.process(imgrgb)
    
    if res.multi_hand_landmarks:
        for hand_landmarks in res.multi_hand_landmarks:
            for counter,loc in enumerate(hand_landmarks.landmark):
                h,w,c=frame.shape
                cx,cy=int(loc.x*w),int(loc.y*h)
                
                
                if cy<h//5:
                    cv2.putText(frame,"please hands off",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
    
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# tashkhise harekate dast rast ya chap

```python
import cv2
import mediapipe as mp

pose=mp.solutions.pose
mppose=pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5)


cap=cv2.VideoCapture(0) # address camera ra midahad, sefr yani webcam
while(True):
    ret,frame=cap.read()
    frame=cv2.flip(frame,1) # drost shodane jahate camera
    imgrgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # tabdile bgr be rgb
    res=mppose.process(imgrgb)
    
    if res.pose_landmarks:
        left_hand_y=res.pose_landmarks.landmark[mp.solutions.pose.PoseLandmark.LEFT_WRIST].y
        right_hand_y=res.pose_landmarks.landmark[mp.solutions.pose.PoseLandmark.RIGHT_WRIST].y
        
        if left_hand_y<right_hand_y:
            cv2.putText(frame,"Right Hand up",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
        else :
            cv2.putText(frame,"Left Hand up",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```
