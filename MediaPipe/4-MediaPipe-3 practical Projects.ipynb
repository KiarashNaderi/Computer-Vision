```python
import cv2
import mediapipe as mp


def is_point_inside_rect(point, rect):
    x, y = point
    x1, y1, x2, y2 = rect
    return x1 < x < x2 and y1 < y < y2


def show_message(frame, message):
    cv2.putText(frame, message, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)

mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.5, min_tracking_confidence=0.5)

cap = cv2.VideoCapture(0)

fruit_pos = (200, 200)
fruit_radius = 20

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = hands.process(frame_rgb)

    cv2.rectangle(frame, (50, 50), (150, 150), (0, 0, 255), 2)

    cv2.rectangle(frame, (400, 400), (500, 500), (0, 255, 0), 2)


    obstacles = [
        {"rect": (200, 200, 250, 250), "color": (255, 0, 0)},
        {"rect": (300, 300, 350, 350), "color": (255, 0, 0)},
        {"rect": (450, 100, 500, 150), "color": (255, 0, 0)}
    ]

    for obstacle in obstacles:
        cv2.rectangle(frame, (obstacle["rect"][0], obstacle["rect"][1]), (obstacle["rect"][2], obstacle["rect"][3]), obstacle["color"], 2)


    cv2.circle(frame, fruit_pos, fruit_radius, (0, 255, 0), -1)


    if results.multi_hand_landmarks:
        hand_landmarks = results.multi_hand_landmarks[0]
        index_finger_tip = (int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * frame.shape[1]),
                            int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * frame.shape[0]))

        
        if is_point_inside_rect(index_finger_tip, (400, 400, 500, 500)):
            show_message(frame, "You won!")

        
        for obstacle in obstacles:
            rect = obstacle["rect"]
            color = obstacle["color"]
            if is_point_inside_rect(index_finger_tip, rect):
                show_message(frame, "You lost!")
                break

        
        if index_finger_tip[0] - fruit_radius < fruit_pos[0] < index_finger_tip[0] + fruit_radius and \
                index_finger_tip[1] - fruit_radius < fruit_pos[1] < index_finger_tip[1] + fruit_radius:
            fruit_pos = index_finger_tip

    
    cv2.imshow('Frame', frame)

    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import random
import time


IMAGE_WIDTH, IMAGE_HEIGHT = 640, 480


score = 0
total_balls = 0
start_time = time.time()
game_duration = 60  


fruit_x = random.randint(50, IMAGE_WIDTH - 50)
fruit_y = IMAGE_HEIGHT - 50


fruit_radius = 20
fruit_speed = 5


fruit_color = (0, 255, 0)


hand_x = 0
hand_y = 0


mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=1,
                       min_detection_confidence=0.7,
                       min_tracking_confidence=0.5)


cap = cv2.VideoCapture(0)
cap.set(3, IMAGE_WIDTH)
cap.set(4, IMAGE_HEIGHT)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    
    elapsed_time = time.time() - start_time
    remaining_time = max(game_duration - int(elapsed_time), 0)


    cv2.putText(frame, f"Time Left: {remaining_time}", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    
    cv2.putText(frame, f"Score: {score}", (20, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    if remaining_time == 0:
        if score > 10:
            result_text = "Winner"
        else:
            result_text = "Loser"
            
        cv2.putText(frame, result_text, (IMAGE_WIDTH // 2 - 100, IMAGE_HEIGHT // 2), cv2.FONT_HERSHEY_SIMPLEX, 2,
                    (0, 0, 255), 4)
        cv2.imshow("Ninja Fruit", frame)
        cv2.waitKey(5000) 
        break


    cv2.circle(frame, (fruit_x, fruit_y), fruit_radius, fruit_color, -1)

    
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(frame_rgb)

    hand_x = 0
    hand_y = 0

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            hand_x = int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * IMAGE_WIDTH)
            hand_y = int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * IMAGE_HEIGHT)

            # برخورد با میوه
            distance = ((hand_x - fruit_x) ** 2 + (hand_y - fruit_y) ** 2) ** 0.5
            if distance < fruit_radius:
                fruit_x = random.randint(50, IMAGE_WIDTH - 50)
                fruit_y = IMAGE_HEIGHT - 50
                score += 1
                total_balls += 1

    
    cv2.circle(frame, (hand_x, hand_y), 15, (255, 0, 0), -1)

    
    fruit_y -= fruit_speed
    if fruit_y < 0:
        fruit_x = random.randint(50, IMAGE_WIDTH - 50)
        fruit_y = IMAGE_HEIGHT - 50

    
    cv2.imshow("Ninja Fruit", frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


final_score = (score / total_balls) * 100 if total_balls > 0 else 0
print(f"Final Score: {final_score:.2f}")

cap.release()
cv2.destroyAllWindows()
```
