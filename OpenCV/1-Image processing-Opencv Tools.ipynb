# change color space

```python
import cv2
img=cv2.imread("Lionel_messi.jpeg")
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.imshow("Lionel1",img)
cv2.imshow("Lionel2",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img=cv2.imread("Lionel_messi.jpeg")
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Lionel1",img)
cv2.imshow("Lionel2",gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(gray_img.shape)
```

# cropping images==> ROI==> Region of interest

```python
#roi==> img_cropped=img[y1:y2,x1:x2]
```

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("Lionel_messi.jpeg")
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

ball=img[280:340,330:390]
ball2=cv2.cvtColor(ball,cv2.COLOR_BGR2RGB)
#plt.imshow(img2)
plt.imshow(ball2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("Lionel_messi.jpeg")

ball=img[280:340,330:390]
img[280:340,80:140]=ball

cv2.imshow("Lionel messi",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# putting texts in images

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("Lionel_messi.jpeg")
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,"Lionel messi",(100,200),font,1,(255,100,180),2)
#(esm file, text morede nazar, location text,font,size font,rang,zekhamat)

cv2.imshow("Lionel messi",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
