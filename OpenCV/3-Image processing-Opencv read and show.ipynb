# read/show videos

```python
import cv2
cap=cv2.VideoCapture(r"E:\my courses\courses\mine\ai\computer vision\Image processing-opencv\data\video\vtest.avi")
while(True):
    ret,frame=cap.read()
    cv2.imshow("Video",frame)
    if cv2.waitKey(1) & 0xFF==ord("q"): #yani karbar q zad filmo beband
        break

cap.release()
cv2.destroyAllWindows()
```

# writing/saving videos

```python
import cv2
cap=cv2.VideoCapture(r"E:\my courses\courses\mine\ai\computer vision\Image processing-opencv\data\video\vtest.avi")
fourcc=cv2.VideoWriter_fourcc("M","P","E","G")
output=cv2.VideoWriter(r"E:\my courses\courses\mine\ai\computer vision\Image processing-opencv\data\video\output.avi",
fourcc,25.0,(768,576))

while(cap.isOpened()): #agar  capture drost ast varede halghe sho
    ret,frame=cap.read() # agar frame  hast true bashe
    if ret==True:
        output.write(frame) # frame be frame zakhire mikonad
        cv2.imshow("video",frame)
        
        if cv2.waitKey(1) & 0xFF==ord("q"):
            break
    else:
        break
cap.release()
output.release()
cv2.destroyAllWindows()
```

# put text on videos

```python
import cv2
cap=cv2.VideoCapture(r"E:\my courses\courses\mine\ai\computer vision\Image processing-opencv\data\video\vtest.avi")


while(cap.isOpened()): 
    ret,frame=cap.read() 
    if ret==True:
        font=cv2.FONT_HERSHEY_SIMPLEX
        frame=cv2.putText(frame,"Never give up",(12,30),font,1,(0,0,0),2)
        
        cv2.imshow("video",frame)

        if cv2.waitKey(1) & 0xFF==ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

# put shape on video

```python
import cv2
cap=cv2.VideoCapture(r"E:\my courses\courses\mine\ai\computer vision\Image processing-opencv\data\video\vtest.avi")


while(cap.isOpened()): 
    ret,frame=cap.read() 
    if ret==True:
        frame=cv2.rectangle(frame,(0,0),(200,250),(255,0,0),5) 
        
        cv2.imshow("video",frame)

        if cv2.waitKey(1) & 0xFF==ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
