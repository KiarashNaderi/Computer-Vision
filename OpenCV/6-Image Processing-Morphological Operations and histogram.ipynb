# erosion ya farsayesh

```python
#cv2.erode()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("Neuron.jpg",0)

ret,thresh=cv2.threshold(img,25,255,cv2.THRESH_BINARY)
kernel=np.ones((3,3),np.uint8)
erroded=cv2.erode(thresh,kernel,iterations=1)

cv2.imshow("original",img)
cv2.imshow("threshold",thresh)
cv2.imshow("erroded",erroded)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

# dilation ya enbesat

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("Neuron.jpg",0)

ret,thresh=cv2.threshold(img,25,255,cv2.THRESH_BINARY)
kernel=np.ones((5,5),np.uint8)
dilated=cv2.dilate(thresh,kernel,iterations=1)

cv2.imshow("original",img)
#cv2.imshow("threshold",thresh)
cv2.imshow("dilated",dilated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# closing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("cells_threshold.png",0)

ret,thresh=cv2.threshold(img,25,255,cv2.THRESH_BINARY)
kernel=np.ones((3,3),np.uint8)
closed=cv2.morphologyEx(thresh,cv2.MORPH_CLOSE,kernel)

cv2.imshow("original",img)
#cv2.imshow("threshold",thresh)
cv2.imshow("closed",closed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# opening

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("cells_threshold.png",0)

ret,thresh=cv2.threshold(img,25,255,cv2.THRESH_BINARY)
kernel=np.ones((15,15),np.uint8)
opened=cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel)

cv2.imshow("original",img)
#cv2.imshow("threshold",thresh)
cv2.imshow("opened",opened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Histogram==> data az brightness va contrast

# histogram tasvir rangi

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("Gyx1k.jpg")
color=('b','g','r')

for i,col in enumerate(color):
    hist=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim([0,256])
    
    
cv2.imshow("image",img)
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# equalized

# clahe()

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("statue.jpg",0)
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8)) # cliplimit daste shoma mibashad
clahe2=clahe.apply(img)

cv2.imshow("image",img)
cv2.imshow("clahe",clahe2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
