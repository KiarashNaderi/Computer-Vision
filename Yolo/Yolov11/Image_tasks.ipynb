```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2


# Load YOLO11 model
model = YOLO("yolo11n.pt")

# Predict and save the result
results = model("/content/drive/MyDrive/Yolo for computer vision/recourses/1.jpg", save=True)

# Extract the predicted image with bounding boxes
img_with_predictions = results[0].plot()

# Display the image using matplotlib
plt.imshow(cv2.cvtColor(img_with_predictions, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.axis("off")  # Hide axes
plt.show()
```

**Pose estimation**

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load a pose estimation model
model = YOLO("yolo11n-pose.pt")

# Predict and display pose estimation
results = model("/content/drive/MyDrive/Yolo for computer vision/recourses/5.jpg", save=True)


img_with_pose = results[0].plot()
plt.imshow(cv2.cvtColor(img_with_pose, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

**Segmentation**

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load a segmentation model
model = YOLO("yolo11n-seg.pt")  # مدل YOLO ویژه تقسیم‌بندی

# Predict and display segmentation
results = model("/content/drive/MyDrive/Yolo for computer vision/recourses/7.jpg", save=True)

# نمایش تصویر با مرزهای تقسیم‌بندی شده
img_with_segmentation = results[0].plot()
plt.imshow(cv2.cvtColor(img_with_segmentation, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
