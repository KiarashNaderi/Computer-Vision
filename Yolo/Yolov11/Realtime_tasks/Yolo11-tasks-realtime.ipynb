# object detection

```python
from ultralytics import YOLO
import cv2

# Load YOLO11 model for object detection
model_detection = YOLO("yolo11n.pt")

# Start video capture from webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Predict and display results
    results = model_detection(frame)
    output_frame = results[0].plot()  # Draw detections on the frame

    cv2.imshow("Object Detection", output_frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Image classification

```python
from ultralytics import YOLO
import cv2

# Load YOLO11 model for image classification
model_classification = YOLO("yolo11n-cls.pt")

# Start video capture from webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Predict the class of the frame
    results = model_classification(frame)
    top_class = results[0].names[results[0].probs.top1]  # Top predicted class
    confidence = results[0].probs.top1conf  # Confidence of top class

    # Display the frame with class label
    cv2.putText(frame, f"{top_class}: {confidence:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("Image Classification", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
